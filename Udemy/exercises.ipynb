### Step 1: Create a Basic Yield Curve from Sample Data. 

Use the following sample U.S. Treasury yields (on-the-run, as of a sample date). These are on-the-run Treasury securities.

```python
maturities = ['3M', '6M', '1Y', '2Y', '5Y', '10Y', '30Y']
yields = [4.5, 4.6, 4.7, 4.6, 4.3, 4.1, 3.9]
```

```python
# Convert maturities to months
maturity_months = [3, 6, 12, 24, 60, 120, 360]

# Create the plot
plt.figure(figsize=(9, 5))
plt.plot(
    maturity_months,
    yields,
    marker='o'
)

# Labels and title
plt.xlabel("Time to Maturity (Months)")
plt.ylabel("Yield (%)")
plt.title("U.S. Treasury On-the-Run Yield Curve")

# Custom x-axis ticks
plt.xticks(maturity_months, maturities)

# Grid for readability
plt.grid(True)

# Display the chart
plt.show()
```

Explanation:
This curve shows an inverted shape, where long-term rates are lower than short-term rates. This is typically interpreted as a signal that investors expect lower interest rates in the future, often due to anticipated economic slowdown.

### Step 2. Compare Yield Curves Across Market Sectors

```python
swap_rates = [4.7, 4.8, 4.9, 5.0, 5.2, 5.3, 5.4]
corp_bond_yields = [5.2, 5.3, 5.4, 5.6, 5.9, 6.1, 6.2]
```

```python
plt.figure(figsize=(9, 5))

plt.plot(maturity_months, yields, marker='o', linewidth=2, color='steelblue', label='U.S. Treasury')
plt.plot(maturity_months, swap_rates, marker='s', linewidth=2, linestyle='--', color='forestgreen', label='Interest Rate Swaps')
plt.plot(maturity_months, corp_bond_yields, marker='^', linewidth=3, linestyle=':', color='indianred', label='Corporate Bonds')

# Labels and title
plt.xlabel("Time to Maturity")
plt.ylabel("Yield (%)")
plt.title("U.S. Treasury, Swap, and Corporate Bond Yield Curves")

# X-axis ticks
plt.xticks(maturity_months, maturities)

# Grid and legend
plt.grid(True)
plt.legend()

plt.show()
```

#### Analysis - comparing curves

Q1: Why is the Treasury yield the lowest among the three?

A1: Treasuries are considered virtually risk-free, backed by the U.S. government, which results in the lowest required yields.

Q2: What risks are priced into swap and corporate bond yields that aren’t in Treasuries?

A2: Swap and corporate yields include credit risk, liquidity risk, and counterparty risk, which raise their yields above Treasuries.

Q3: How does liquidity affect the shape and position of each yield curve?

A3: More liquid markets like Treasuries have flatter or more stable curves. Less liquid assets require higher yields to attract investors, making their curves steeper.

Q4: What would it imply if the corporate yield curve suddenly became inverted?

A4: It may signal higher perceived default risk or future financial stress in the corporate sector, or it could reflect expectations of lower future interest rates.
